````python
def mean_filter(_img, K_size=3):
    
    pad = K_size//2
    H, W, C = _img.shape
    
    img = np.zeros((H+2*pad, W+2*pad, C), dtype=np.float)
    img[pad:H+pad, pad:W+pad] = _img.copy().astype(np.float)
    out = np.zeros_like(_img)
    
    for y in range(H):
        for x in range(W):
            for c in range(C):
                out[y, x, c] = np.mean(img[y:y+K_size, x:x+K_size, c])
    
    return out.astype(np.uint8)
````

````python
def motion_filter(_img, K_size=3):
    pad = K_size//2
    H, W, C = _img.shape
    
    img = np.zeros((H+2*pad, W+2*pad, C), dtype=np.float)
    img[pad:H+pad, pad:W+pad] = _img.copy().astype(np.float)
    out = np.zeros_like(_img)
    
    for y in range(H):
        for x in range(W):
            for c in range(C):
                for i in range(K_size):
                    out[y, x, c] += img[y+i, x+i, c] / K_size
    
    return out.astype(np.uint8)
````

````python
def max_min_filter(img_, K_size=3):
    _img = np.array(img_.copy())
    if len(_img.shape) == 3:
        H, W, C = _img.shape
    else:
        H, W = _img.shape
        C = 1
        _img = _img.reshape(H, W, C)
    
    pad = K_size//2
    img = np.zeros((H+2*pad, W+2*pad, C), dtype=np.float)
    img[pad:H+pad, pad:W+pad] = _img.copy().astype(np.float)
    out = np.zeros_like(_img)
    
    for y in range(H):
        for x in range(W):
            for c in range(C):
                Max = np.max(img[y:y+K_size, x:x+K_size, c])
                Min = np.min(img[y:y+K_size, x:x+K_size, c])
                out[y, x, c] = Max-Min
    if C == 1:
        out = out.reshape(H, W)
    return out.astype(np.uint8)
````

````python
def differential_filter(img_, K_size=3):
    _img = np.array(img_.copy())
    if len(_img.shape) == 3:
        H, W, C = _img.shape
    else:
        H, W = _img.shape
        C = 1
        _img = _img.reshape(H, W, C)
    pad = K_size//2
    img = np.zeros((H+2*pad, W+2*pad, C), dtype=np.float)
    img[pad:H+pad, pad:W+pad] = _img.copy().astype(np.float)
    out_v = np.zeros_like(_img)
    out_h = np.zeros_like(_img)
    k_v = np.zeros((K_size, K_size)).copy()
    k_h = np.zeros((K_size, K_size)).copy()
    k_v[0, 1] = -1
    k_v[1, 1] = 1
    k_h[1, 0] = -1
    k_h[1, 1] = 1
    
    for y in range(H):
        for x in range(W):
            for c in range(C):
                out_v[y, x, c] = np.abs(np.sum(k_v* img[y: y+K_size, x: x+K_size, c]))
                out_h[y, x, c] = np.abs(np.sum(k_h* img[y: y+K_size, x: x+K_size, c]))
    out = np.clip(out_v, 0, 255)
    if C==1:
        out_v = out_v.reshape(H, W)
        out_h = out_h.reshape(H, W)
    return out_v.astype(np.uint8), out_h.astype(np.uint8)
````

````python
def prewitt_filter(img_, K_size=3):
    _img = np.array(img_.copy())
    if len(_img.shape) == 3:
        H, W, C = _img.shape
    else:
        H, W = _img.shape
        C = 1
        _img = _img.reshape(H, W, C)
    pad = K_size//2
    img = np.zeros((H+2*pad, W+2*pad, C), dtype=np.float)
    img[pad:H+pad, pad:W+pad] = _img.copy().astype(np.float)
    out_v = np.zeros_like(_img)
    out_h = np.zeros_like(_img)
    k_v = np.zeros((K_size, K_size)).copy()
    k_h = np.zeros((K_size, K_size)).copy()
    k_v[0, :] = -1
    k_v[K_size-1, :] = 1
    k_h[:, 0] = -1
    k_h[:, K_size-1] = 1
    
    for y in range(H):
        for x in range(W):
            for c in range(C):
                out_v[y, x, c] = max(0, np.sum(k_v* img[y: y+K_size, x: x+K_size, c]))
                out_h[y, x, c] = max(0, np.sum(k_h* img[y: y+K_size, x: x+K_size, c]))
    if C==1:
        out_v = out_v.reshape(H, W)
        out_h = out_h.reshape(H, W)
    return out_v.astype(np.uint8), out_h.astype(np.uint8)
````

````python
def sobel_filter(img_, K_size=3):
    _img = np.array(img_.copy())
    if len(_img.shape) == 3:
        H, W, C = _img.shape
    else:
        H, W = _img.shape
        C = 1
        _img = _img.reshape(H, W, C)
    pad = K_size//2
    img = np.zeros((H+2*pad, W+2*pad, C), dtype=np.float)
    img[pad:H+pad, pad:W+pad] = _img.copy().astype(np.float)
    out_v = np.zeros_like(_img)
    out_h = np.zeros_like(_img)
    k_v = np.zeros((K_size, K_size), dtype=np.float)
    k_h = np.zeros((K_size, K_size), dtype=np.float)
    for i in range(K_size):
        k_v[0, i] = min(i+1, K_size-i)
        k_v[K_size-1, i] = -min(i+1, K_size-i)
        k_h[i, 0] = min(i+1, K_size-i)
        k_h[i, K_size-1] = -min(i+1, K_size-i)
    for y in range(H):
        for x in range(W):
            for c in range(C):
                out_v[y, x, c] = max(0.0, np.sum(k_v* img[y: y+K_size, x: x+K_size, c]))
                out_h[y, x, c] = max(0.0, np.sum(k_h* img[y: y+K_size, x: x+K_size, c]))
    if C==1:
        out_v = out_v.reshape(H, W)
        out_h = out_h.reshape(H, W)
    return out_v.astype(np.uint8), out_h.astype(np.uint8)
````

````python
def laplacian_filter(img_, K_size=3):
    K_size=3
    _img = np.array(img_.copy())
    if len(_img.shape) == 3:
        H, W, C = _img.shape
    else:
        H, W = _img.shape
        C = 1
        _img = _img.reshape(H, W, C)
    pad = K_size//2
    img = np.zeros((H+2*pad, W+2*pad, C), dtype=np.float)
    img[pad:H+pad, pad:W+pad] = _img.copy().astype(np.float)
    out = np.zeros_like(_img)
    k = np.array([[0,1,0], [1,-4,1], [0,1,0]])
    for y in range(H):
        for x in range(W):
            for c in range(C):
                out[y, x, c] = max(0, np.sum(k* img[y: y+K_size, x: x+K_size, c]))
    if C==1:
        out = out.reshape(H, W)
    return out.astype(np.uint8)
````

````python
def LoG_filter(img_, sigma=3, K_size=5):
    _img = np.array(img_.copy())
    if len(_img.shape) == 3:
        H, W, C = _img.shape
    else:
        H, W = _img.shape
        C = 1
        _img = _img.reshape(H, W, C)
    pad = K_size//2
    img = np.zeros((H+2*pad, W+2*pad, C), dtype=np.float)
    img[pad:H+pad, pad:W+pad] = _img.copy().astype(np.float)
    out = np.zeros_like(_img)
    k = np.zeros((K_size, K_size), dtype=np.float)
    for y in range(K_size):
        for x in range(K_size):
            i, j = y-pad, x-pad
            k[y, x] = i**2 + j**2 - 2*(sigma**2) #
            k[y, x] /= 2 * np.pi * (sigma**6)    #
            k[y, x] *= np.exp(-(i**2 + i**2) / (2*(sigma**2)))
    k /= k.sum()
    print(k)
    for y in range(H):
        for x in range(W):
            for c in range(C):
                out[y, x, c] = min(255, max(0, np.sum(k* img[y: y+K_size, x: x+K_size, c])))
    if C==1:
        out = out.reshape(H, W)
    return out.astype(np.uint8)
````
